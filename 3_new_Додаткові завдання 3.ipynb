# –î–æ–¥–∞—Ç–∫–æ–≤—ñ –∑–∞–≤–¥–∞–Ω–Ω—è ‚Ññ3

## 1 –ü–µ—Ä–µ–≤—ñ—Ä–∫–∞ –≤–∑–∞—î–º–Ω–æ—ó –ø—Ä–æ—Å—Ç–æ—Ç–∏ –¥–≤–æ—Ö —á–∏—Å–µ–ª
–†–µ–∞–ª—ñ–∑—É–π—Ç–µ –∞–ª–≥–æ—Ä–∏—Ç–º –ï–≤–∫–ª—ñ–¥–∞ –¥–ª—è –ø–µ—Ä–µ–≤—ñ—Ä–∫–∏ –≤–∑–∞—î–º–Ω–æ—ó –ø—Ä–æ—Å—Ç–æ—Ç–∏ –¥–≤–æ—Ö —á–∏—Å–µ–ª (—É –≤–∏–≥–ª—è–¥—ñ —Ñ—É–Ω–∫—Ü—ñ—ó –≤—ñ–¥ –¥–≤–æ—Ö –∞—Ä–≥—É–º–µ–Ω—Ç—ñ–≤).
–í–∏–∫–æ–Ω–∞–π—Ç–µ –¥—Ä—É–≥–µ –∑–∞–≤–¥–∞–Ω–Ω—è –∑ **–î–æ–º–∞—à–Ω—å–æ—ó —Ä–æ–±–æ—Ç–∏ ‚Ññ3** (**–ó–∞–¥–∞—á–∞ 1.3.14**) –∑ –≤–∏–∫–æ—Ä–∏—Å—Ç–∞–Ω–Ω—è–º –¥–∞–Ω–æ—ó —Ñ—É–Ω–∫—Ü—ñ—ó.

```python
# –ú—ñ—Å—Ü–µ –¥–ª—è –í–∞—à–æ–≥–æ –∫–æ–¥—É
from itertools import combinations
import numpy as np
def Evklid (a, b):
    x=a
    y=b
    while y!=0:
        x, y=y, x%y
    print(x)
    return(x==1)


def teor (n, k, p, q):
    if Evklid (p,q):
        my_range=range(1,n+1)
        choose_k = list(combinations(my_range,k))
        n1, n2, n3=0, 0, 0
        d=len(choose_k)
        for i in choose_k:
            Flag1=True
            Flag2=True
            Flag3=False
            for j in i:
                if j%p==0:
                    Flag3=True
                elif j%q==0:
                    Flag1=False
                else:
                    Flag1=False
                    Flag2=False
            n1+=Flag1
            n2+=Flag2
            n3+=Flag3
        print ('–∫–æ–∂–Ω–µ –∑ –≤–∏–±—Ä–∞–Ω–∏—Ö —á–∏—Å–µ–ª –∫—Ä–∞—Ç–Ω–µ –¥–∞–Ω–æ–º—É —á–∏—Å–ª—É  ùëù: {};  –∫–æ–∂–Ω–µ –∑ –≤–∏–±—Ä–∞–Ω–∏—Ö —á–∏—Å–µ–ª –∫—Ä–∞—Ç–Ω–µ —Ö–æ—á–∞ –± –æ–¥–Ω–æ–º—É –∑ –¥–≤–æ—Ö –≤–∑–∞—î–º–Ω–æ –ø—Ä–æ—Å—Ç–∏—Ö —á–∏—Å–µ–ª  ùëù  i  ùëû :{}; —Å–µ—Ä–µ–¥ –≤–∏–±—Ä–∞–Ω–∏—Ö —á–∏—Å–µ–ª —î —Ö–æ—á–∞ –± –æ–¥–Ω–µ –∫—Ä–∞—Ç–Ω–µ  ùëù :{}'.format(n1/d, n2/d, n3/d)) 
    else:
        print(p, q, '–Ω–µ –≤–∑–∞—î–º–Ω–æ–ø—Ä–æ—Å—Ç—ñ')
teor(ùëõ=25,ùëò=5,ùëù=3,ùëû=4) 
teor(ùëõ=25,ùëò=10,ùëù=3,ùëû=4) 
teor(ùëõ=25,ùëò=10,ùëù=30,ùëû=40) 

def emp (n, k, p, q):
    if Evklid (p,q):
        my_range=range(1,n+1)
        choose_k =[]
        for i in range(10000):
            choose_k.append(list(np.random.choice(my_range, size=k, replace=False)))
        n1, n2, n3=0, 0, 0
        d=len(choose_k)
        for i in choose_k:
            Flag1=True
            Flag2=True
            Flag3=False
            for j in i:
                if j%p==0:
                    Flag3=True
                elif j%q==0:
                    Flag1=False
                else:
                    Flag1=False
                    Flag2=False
            n1+=Flag1
            n2+=Flag2
            n3+=Flag3
        print ('–∫–æ–∂–Ω–µ –∑ –≤–∏–±—Ä–∞–Ω–∏—Ö —á–∏—Å–µ–ª –∫—Ä–∞—Ç–Ω–µ –¥–∞–Ω–æ–º—É —á–∏—Å–ª—É  ùëù: {};  –∫–æ–∂–Ω–µ –∑ –≤–∏–±—Ä–∞–Ω–∏—Ö —á–∏—Å–µ–ª –∫—Ä–∞—Ç–Ω–µ —Ö–æ—á–∞ –± –æ–¥–Ω–æ–º—É –∑ –¥–≤–æ—Ö –≤–∑–∞—î–º–Ω–æ –ø—Ä–æ—Å—Ç–∏—Ö —á–∏—Å–µ–ª  ùëù  i  ùëû :{}; —Å–µ—Ä–µ–¥ –≤–∏–±—Ä–∞–Ω–∏—Ö —á–∏—Å–µ–ª —î —Ö–æ—á–∞ –± –æ–¥–Ω–µ –∫—Ä–∞—Ç–Ω–µ  ùëù :{}'.format(n1/d, n2/d, n3/d)) 
    else:
        print(p, q, '–Ω–µ –≤–∑–∞—î–º–Ω–æ–ø—Ä–æ—Å—Ç—ñ')
emp(ùëõ=25,ùëò=5,ùëù=3,ùëû=4) 
emp(ùëõ=25,ùëò=10,ùëù=3,ùëû=4) 
emp(ùëõ=25,ùëò=10,ùëù=30,ùëû=40) 
```

## 2 –ê–Ω–∞–ª—ñ–∑ —Ñ—É–Ω–∫—Ü—ñ–π –∑ –∫–ª–∞—Å—É *numpy.random*
–ü—Ä–æ–∞–Ω–∞–ª—ñ–∑—É–π—Ç–µ –≤–∑–∞—î–º–æ–∑–≤'—è–∑–æ–∫ –º—ñ–∂ —Ñ—É–Ω–∫—Ü—ñ—è–º–∏ –∑ –∫–ª–∞—Å—É **numpy.random**:

```python
# –ú—ñ—Å—Ü–µ –¥–ª—è –í–∞—à–æ–≥–æ –∫–æ–¥—É —Ç–∞ –≤–∏—Å–Ω–æ–≤–∫—ñ–≤
import numpy.random
print ('rand', numpy.random.rand (2, 3, 4)) #–ø–æ–≤–µ—Ä—Ç–∞—î –º–∞—Å–∏–≤ —Ä–æ–∑–º—ñ—Ä—É 2*3*4 (–∞—Ä–≥—É–º–µ–Ω—Ç–∏ —Ñ—É–Ω–∫—Ü—ñ—ó) - –∑ —á–∏—Å–µ–ª –≤—ñ–¥ 0 –¥–æ 1
x1=numpy.random.randint (1, 10, size=(2, 3, 4))
print ('randint', type(x1[0, 0, 0]), x1) #–ø–æ–≤–µ—Ä—Ç–∞—î –º–∞—Å–∏–≤ —Ä–æ–∑–º—ñ—Ä—É 2*3*4 (size) - –∑ —á–∏—Å–µ–ª –≤—ñ–¥ 1 –¥–æ 9 (–≤—ñ–¥ –ø–µ—Ä—à–æ–≥–æ –∞—Ä–≥—É–º–µ–Ω—Ç—É –¥–æ –¥—Ä—É–≥–æ–≥–æ)
x2=numpy.random.random_integers (1, 10, size=(2, 3, 4))# —Ä–æ–±–∏—Ç—å —Ç–µ–∂ —Å–∞–º–µ, –∞–ª–µ –¥—Ä—É–≥–∏–π –ø–∞—Ä–∞–º–µ—Ç—Ä –≤–∫–ª—é—á–Ω–æ, –∞ —è–∫—â–æ –¥—Ä—É–≥–æ–≥–æ –ø–∞—Ä–∞–º–µ—Ç—Ä—É –Ω–µ–º–∞, –±–µ—Ä–µ —á–∏—Å–ª–∞ –≤—ñ–¥ 1 –¥–æ –ø–µ—Ä—à–æ–≥–æ –ø–∞—Ä–∞–º–µ—Ç—Ä—É(–≤–∫–ª—é—á–Ω–æ. 

print ('random_integers', type(x2[0, 0, 0]), x2) #–ø–æ–≤–µ—Ä—Ç–∞—î –º–∞—Å–∏–≤ —Ä–æ–∑–º—ñ—Ä—É 2*3*4 (size) - –∑ —á–∏—Å–µ–ª –≤—ñ–¥ 1 –¥–æ 9 (–≤—ñ–¥ –ø–µ—Ä—à–æ–≥–æ –∞—Ä–≥—É–º–µ–Ω—Ç—É –¥–æ –¥—Ä—É–≥–æ–≥–æ)
shape=(2, 3, 4)
print('sample', np.random.random_sample(shape))#–ø–æ–¥—ñ–±–Ω–∞ –¥–æ random.rand, –∞–ª–µ –ø—Ä–∏–π–º–∞—î –≤ —è–∫–æ—Å—Ç—ñ –ø–∞—Ä–∞–º–µ—Ç—Ä—ñ–≤ –Ω–µ –¥–æ–≤—ñ–ª—å–Ω—É –∫—ñ–ª—å–∫—ñ—Å—Ç—å —Ü—ñ–ª–∏—Ö —á–∏—Å–µ–ª, –∞ –æ–¥—Ä–∞–∑—É –∫–æ—Ä—Ç–µ–∂ –∑ –Ω–∏—Ö
shape=(2, 3, 4)
print('random', np.random.random(shape))#–ï–∫–≤—ñ–≤–∞–ª–µ–Ω—Ç–Ω–∞ –¥–æ np.random.random_sample. –ü–æ–¥—ñ–±–Ω–∞ –¥–æ random.rand, –∞–ª–µ –ø—Ä–∏–π–º–∞—î –≤ —è–∫–æ—Å—Ç—ñ –ø–∞—Ä–∞–º–µ—Ç—Ä—ñ–≤ –Ω–µ –¥–æ–≤—ñ–ª—å–Ω—É –∫—ñ–ª—å–∫—ñ—Å—Ç—å —Ü—ñ–ª–∏—Ö —á–∏—Å–µ–ª, –∞ –æ–¥—Ä–∞–∑—É –∫–æ—Ä—Ç–µ–∂ –∑ –Ω–∏—Ö
shape=(2, 3, 4)
print('ranf', np.random.ranf(shape))#–ï–∫–≤—ñ–≤–∞–ª–µ–Ω—Ç–Ω–∞ –¥–æ np.random.random_sample. –ü–æ–¥—ñ–±–Ω–∞ –¥–æ random.rand, –∞–ª–µ –ø—Ä–∏–π–º–∞—î –≤ —è–∫–æ—Å—Ç—ñ –ø–∞—Ä–∞–º–µ—Ç—Ä—ñ–≤ –Ω–µ –¥–æ–≤—ñ–ª—å–Ω—É –∫—ñ–ª—å–∫—ñ—Å—Ç—å —Ü—ñ–ª–∏—Ö —á–∏—Å–µ–ª, –∞ –æ–¥—Ä–∞–∑—É –∫–æ—Ä—Ç–µ–∂ –∑ –Ω–∏—Ö
shape=(2, 3, 4)
print('sample', np.random.sample(shape))#–ï–∫–≤—ñ–≤–∞–ª–µ–Ω—Ç–Ω–∞ –¥–æ np.random.random_sample. –ü–æ–¥—ñ–±–Ω–∞ –¥–æ random.rand, –∞–ª–µ –ø—Ä–∏–π–º–∞—î –≤ —è–∫–æ—Å—Ç—ñ –ø–∞—Ä–∞–º–µ—Ç—Ä—ñ–≤ –Ω–µ –¥–æ–≤—ñ–ª—å–Ω—É –∫—ñ–ª—å–∫—ñ—Å—Ç—å —Ü—ñ–ª–∏—Ö —á–∏—Å–µ–ª, –∞ –æ–¥—Ä–∞–∑—É –∫–æ—Ä—Ç–µ–∂ –∑ –Ω–∏—Ö
shape=(2, 3, 4)
A=range(10)
p=[0.02, 0.1, 0.05, 0.05, 0.2, 0.15, 0.025, 0.025, 0.37, 0.01]
print('choice', np.random.choice(A, shape, replace=True, p=p))#–û–±–∏—Ä–∞—î –∑ –ø–µ—Ä—à–æ–≥–æ –∞—Ä–≥—É–º–µ–Ω—Ç—É - –æ–±'—î–∫—Ç–∞ —Ç–∏–ø—É, —â–æ —ñ—Ç–µ—Ä—É—î—Ç—å—Å—è, –º–∞—Å–∏–≤ —Ä–æ–∑–º—ñ—Ä—ñ–≤ shape.
#–Ø–∫—â–æ replace=True, —Ç–æ –æ–±'—î–∫—Ç–∏ –∑ –ê –º–æ–∂—É—Ç—å –≤—Ö–æ–¥–∏—Ç–∏ –≤ –Ω–æ–≤–∏–π –º–∞—Å–∏–≤ –¥–µ–∫—ñ–ª—å–∫–∞ —Ä–∞–∑—ñ–≤. –û—Å—Ç–∞–Ω–Ω—ñ–π –∞—Ä–≥—É–º–µ–Ω—Ç - —Ü–µ –π–º–æ–≤—ñ—Ä–Ω–æ—Å—Ç—ñ –∫–æ–∂–Ω–æ—ó –∑ –ø–æ–¥—ñ–π –¥–æ–≤–∂–∏–Ω–∞ —Ä –º–∞—î –¥–æ—Ä—ñ–≤–Ω—é–≤–∞—Ç–∏ –¥–æ–≤–∂–∏–Ω—ñ –ê
arr = np.arange(10)
print('before np.random.shuffle'+str(arr))
(np.random.shuffle(arr))#–ø–µ—Ä–µ–º—ñ—à—É—î –ø–µ—Ä—à–∏–π –∞—Ä–≥—É–º–µ–Ω—Ç - –æ–±'—î–∫—Ç —Ç–∏–ø—É, —â–æ —ñ—Ç–µ—Ä—É—î—Ç—å—Å—è. –ù—ñ—á–æ–≥–æ –Ω–µ –ø–æ–≤–µ—Ä—Ç–∞—î, –∞–ª–µ arr –≤–∂–µ –ø–µ—Ä–µ–º—ñ—à–∞–Ω–∏–π
print('after np.random.shuffle'+str(arr))
print(numpy.random.permutation(7))#—è–∫—â–æ –∞—Ä–≥—É–º–µ–Ω—Ç numpy.random.permutation - —á–∏—Å–ª–æ, —Ç–æ –ø–æ–≤–µ—Ä—Ç–∞—î –ø–µ—Ä–µ–º—ñ—à–∞–Ω—É –ø–æ—Å–ª—ñ–¥–æ–≤–Ω—ñ—Å—Ç—å –≤—ñ–¥ 0 –¥–æ –∞—Ä–≥—É–º–µ–Ω—Ç—É (–Ω–µ –≤–∫–ª—é—á–Ω–æ)
a=[1,7,2]
print('numpy.random.permutation', numpy.random.permutation(a)) #—è–∫—â–æ –∞—Ä–≥—É–º–µ–Ω—Ç numpy.random.permutation - –æ–±'—î–∫—Ç–∞ —Ç–∏–ø—É, —â–æ —ñ—Ç–µ—Ä—É—î—Ç—å—Å—è, —Ç–æ –ø–æ–≤–µ—Ä—Ç–∞—î –ø–µ—Ä–µ–º—ñ—à–∞–Ω—É –ö–û–ü–Ü–Æ —Ü—å–æ–≥–æ –æ–±'—î–∫—Ç—É

print(a)
```
