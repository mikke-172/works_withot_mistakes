### Завдання 1
Реалізуйте випадкову величину, описану в умові задачі, за допомогою класу [scipy.stats.rv_discrete](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.rv_discrete.html) в залежності від параметра $n$.

Створіть дві випадкові величини для:
- $n = 10$;
- $n = 25$.

```python
# Місце для Вашого коду
import numpy as np
import math
from scipy.stats import rv_discrete
def do (n):
    s=0
    for i in range (1, 2*n+1):
        s+=(math.log(i))
    k=1/s
    l=[]
    for i in range (1, 2*n+1):
        l.append(k*math.log(i))
    return(l)
n1=np.array(do(10))
n2=np.array(do(25))
xi1=rv_discrete (a=1, b=20, values=(np.arange(1, 21), n1))
xi2=rv_discrete (a=1, b=20, values=(np.arange(1, 51), n2))
```

### Завдання 2
Зобразіть графічно функцію щільності, функцію розподілу та гістограму $10000$ значень отриманої випадкової величини в залежності від параметра $n$:

```python
# Місце для Вашого коду
import matplotlib.pyplot as plt
x1=np.arange(0, 21, 1)
y1 = xi1.pmf(x1)
xi1_rvs=xi1.rvs(size=10000)
print(xi1_rvs)
plt.stem(x1, y1)
plt.show()
y1= xi1.cdf(x1)
plt.plot(x1, y1)
plt.show()
plt.hist(xi1_rvs)
plt.show()
x2=np.arange(0, 51, 1)
y2 = xi2.pmf(x2)
xi2_rvs=xi2.rvs(size=10000)
plt.stem(x2, y2)
plt.show()
y1= xi1.cdf(x2)
plt.plot(x2, y2)
plt.show()
plt.hist(xi2_rvs)
plt.show()
```
